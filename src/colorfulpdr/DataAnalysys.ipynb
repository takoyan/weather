# import module

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import glob
import os
```

# load module

```python
datas=pd.read_csv('../../etc/dif_features/clipped_std_features.csv')
moving=datas[datas['label']=='m']
observing=datas[datas['label']=='o']
searching=datas[datas['label']=='s']
```

```python
columns=datas.columns[:len(datas.columns)-2]
```

# plot

```python
for column in columns:
    fig, ax = plt.subplots()

    bp = ax.boxplot((moving[column], observing[column], searching[column]))
    ax.set_xticklabels(['M'+column, 'O'+column, 'S'+column])

    plt.title('Box plot')
    plt.xlabel('features')
    plt.ylabel('score')
    # Y軸のメモリのrange
    #plt.ylim([0,100])
    plt.grid()

    # fig.savefig("../../etc/picture/clipped_std/{}.png".format(column))

    # 描画
    plt.show()
```
